```python
import numpy as np
import matplotlib.pyplot as plt

# Possible curve segment types
def clamp(x, min_v, max_v):
    return max(min_v, min(max_v, x))

def ease_constant(x: float, start: float, end: float, v: float = 1.0):
    if start <= x <= end:
        return v
    else:
        return 0

def ease_linear(x: float, min_t: float, max_t: float, min_v: float = 0.0, max_v: float = 1.0):
    t = (x - min(min_t, max_t)) / abs(max_t - min_t)
    if t < 0 or t > 1:
        return 0
    
    if min_t > max_t:
        t = 1 - t

    return (min_v * (1 - t)) + (max_v * t)

def ease_in_out(x: float, min_t: float, max_t: float, min_v: float = 0.0, max_v: float = 1.0):
    t = (x - min(min_t, max_t)) / abs(max_t - min_t)
    if t < 0 or t > 1:
        return 0

    if min_t > max_t:
        t = 1 - t

    if t < 0.5:
        t = 4 * (t ** 3)
    else:
        t = 1 - ((-2 * t + 2) ** 3) / 2
        
    return (min_v * (1 - t)) + (max_v * t)
```

```python
def sample_a1_curve_a(x: float):
    v = 0
    v = max(v, ease_in_out(x, 10, 22.5))
    v = max(v, ease_constant(x, 22.5, 27.5, v = 1))
    v = max(v, ease_in_out(x, 40, 27.5))

    
    v = max(v, ease_in_out(x, 20, 80, min_v=0.1))
    v = max(v, ease_constant(x, 80, 90, v = 1))
    v = max(v, ease_in_out(x, 100, 90))
    return v

def sample_a1_curve_b(x: float):
    v = 0
    v = max(v, ease_linear(x, 10, 22.5))
    v = max(v, ease_constant(x, 22.5, 27.5, v = 1))
    v = max(v, ease_linear(x, 40, 27.5))

    
    v = max(v, ease_linear(x, 20, 80, min_v=0.1))
    v = max(v, ease_constant(x, 80, 90, v = 1))
    v = max(v, ease_linear(x, 100, 90))
    return v

a1_x = list(range(0, 151))
a1_ys = []
a1_ys.append([sample_a1_curve_a(x) * 100 for x in a1_x])
a1_ys.append([sample_a1_curve_b(x) * 100 for x in a1_x])

# for a1_y in a1_ys:
#     plt.plot(a1_x, a1_y)
#     plt.xlabel('A1')
#     plt.ylabel('Selection probability (%)')
#     plt.xticks(np.arange(0, 151, 10))
#     plt.grid()
#     plt.tight_layout()
#     plt.show()

fig, axs = plt.subplots(ncols=len(a1_ys), layout='constrained', figsize=(15, 5))

for ax, a1_y, label in zip(axs, a1_ys, ['A1 (a)', 'A1 (b)']):
    ax.plot(a1_x, a1_y)
    ax.set_xlabel(label)
    ax.set_ylabel('Selection probability (%)')
    ax.set_xticks(np.arange(0, 151, 10))
    ax.grid()

plt.tight_layout()
fig.show()
```

```python
def sample_a2_curve_a(x: float):
    v = 0
    v = max(v, ease_in_out(x, 0, 20))
    v = max(v, ease_constant(x, 20, 30))
    v = max(v, ease_in_out(x, 100, 30))
    return v

def sample_a2_curve_b(x: float):
    v = 0
    v = max(v, ease_linear(x, 0, 20))
    v = max(v, ease_constant(x, 20, 30))
    v = max(v, ease_linear(x, 100, 30))
    return v

a2_x = list(range(0, 101))
a2_ys = []
a2_ys.append([sample_a2_curve_a(x) * 100 for x in a2_x])
a2_ys.append([sample_a2_curve_b(x) * 100 for x in a2_x])

# for a2_y in a2_ys:
#     plt.plot(a2_x, a2_y)
#     plt.xlabel('A2')
#     plt.ylabel('Selection probability (%)')
#     plt.xticks(np.arange(0, 101, 10))
#     plt.grid()
#     plt.tight_layout()
#     plt.show()

fig, axs = plt.subplots(ncols=len(a2_ys), layout='constrained', figsize=(15, 5))

for ax, a2_y, label in zip(axs, a2_ys, ['A2 (a)', 'A2 (b)']):
    ax.plot(a2_x, a2_y)
    ax.set_xlabel(label)
    ax.set_ylabel('Selection probability (%)')
    ax.set_xticks(np.arange(0, 101, 10))
    ax.grid()

plt.tight_layout()
fig.show()
```

```python
num_points = 2000

import os
import json
import random

def create_a1_value():
    return random.uniform(0, 150)

def create_a2_value():
    return random.uniform(0, 100)

def compute_selection_probability(a1: float, a2: float):
    p1 = sample_a1_curve_a(a1)
    p2 = sample_a2_curve_a(a2)

    return p1 * p2

def compute_class(a1: float, a2: float):
    prob = compute_selection_probability(a1, a2)

    if 0.25 <= prob <= 0.75:
        return 0.75
    else:
        return 0.25

def create_dataset():
    labels = [i for i in range(num_points)]
    a1 = [create_a1_value() for _ in range(num_points)]
    a2 = [create_a2_value() for _ in range(num_points)]
    cl = [compute_class(a1, a2) for a1, a2 in zip(a1, a2)]

    return {
        'a1': a1,
        'a2': a2,
        'labels': labels,
        'cl': cl,
    }

dataset = create_dataset()
obj = {
    'a1': {
        'label': 'A1',
        'range': [0, 150],
        'dataPoints': dataset['a1'],
        'tickPositions': list(map(lambda x: int(x), np.arange(0, 151, 10))),
        'tickLabels': list(map(lambda x: str(x), np.arange(0, 151, 10)))
    },
    'a2': {
        'label': 'A2',
        'range': [0, 100],
        'dataPoints': dataset['a2'],
        'tickPositions': list(map(lambda x: int(x), np.arange(0, 101, 10))),
        'tickLabels': list(map(lambda x: str(x), np.arange(0, 101, 10)))
    },
    'label': {
        'label': 'Label',
        'dataPoints': dataset['labels']
    },
    'class': {
        'label': 'Class',
        'range': [0, 1],
        'tickPositions': [0.25, 0.75],
        'tickLabels': ['C2', 'C1'],
        'dataPoints': dataset['cl']
    }
}

print(json.dumps(obj, indent=2))
```
