```python
import numpy as np
import matplotlib.pyplot as plt

# Possible curve segment types
def clamp(x, min_v, max_v):
    return max(min_v, min(max_v, x))

def ease_constant(x: float, start: float, end: float):
    if start <= x <= end:
        return 1
    else:
        return 0

def ease_linear(x: float, min_v: float, max_v: float):
    x = (x - min(min_v, max_v)) / abs(max_v - min_v)
    if x < 0 or x > 1:
        return 0
    
    if min_v <= max_v:
        return x
    else:
        return 1 - x

def ease_in_out(x: float, min_v: float, max_v: float):
    x = (x - min(min_v, max_v)) / abs(max_v - min_v)
    if x < 0 or x > 1:
        return 0

    if min_v > max_v:
        x = 1 - x

    if x < 0.5:
        return 4 * (x ** 3)
    else:
        return 1 - ((-2 * x + 2) ** 3) / 2
```

```python
def sample_a1_curve(x: float):
    v = 0
    v = max(v, ease_in_out(x, 50, 125))
    v = max(v, ease_constant(x, 125, 175))
    v = max(v, ease_in_out(x, 250, 175))
    v = max(v, ease_in_out(x, 250, 325))
    v = max(v, ease_constant(x, 325, 375))
    v = max(v, ease_in_out(x, 450, 375))
    return v

a1_x = list(range(0, 501))
a1_y = [sample_a1_curve(x) * 100 for x in a1_x]

plt.plot(a1_x, a1_y)
plt.xlabel('A1')
plt.ylabel('Selection probability (%)')
plt.xticks(np.arange(0, 501, 50))
plt.grid()
plt.tight_layout()
plt.show()
```

```python
def sample_a2_curve(x: float):
    v = 0
    v = max(v, ease_in_out(x, 100, 170))
    v = max(v, ease_constant(x, 170, 180))
    v = max(v, ease_in_out(x, 200, 180))

    return v

a2_x = list(range(100, 201))
a2_y = [sample_a2_curve(x) * 100 for x in a2_x]

plt.plot(a2_x, a2_y)
plt.xlabel('A2')
plt.ylabel('Selection probability (%)')
plt.xticks(np.arange(100, 201, 10))
plt.grid()
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, layout='constrained', figsize=(15, 5))

ax1.plot(a1_x, a1_y)
ax1.set_xlabel('A1')
ax1.set_ylabel('Selection probability (%)')
ax1.set_xticks(np.arange(0, 501, 50))
ax1.grid()

ax2.plot(a2_x, a2_y)
ax2.set_xlabel('A2')
ax2.set_ylabel('Selection probability (%)')
ax2.set_xticks(np.arange(100, 201, 10))
ax2.grid()

plt.tight_layout()
fig.show()
```
